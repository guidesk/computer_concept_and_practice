# 집합 자료형 생성
---
중괄호 {} 사용
* a_set = {item1, item2, ..., itemN}
* {} 는 딕셔너리 자료형이 생성됨.
* set() 함수 사용
* 다른 시퀀스 자료형으로의 변환 : list(), tuple(), str() 으로 가능
* Mutable 시퀀스 타입
    * 리스트 자료형과 동일
    * 튜플, 문자열 자료형은 immutable 시퀀스 타입
* 중복 아이템들은 제거되고, 유일한 아이템들만 남는다.
* 순서가 지정되지 않음!!
    * 순서를 이용해서 가능한 인덱싱, 슬라이싱 등은 사용 불가.

# Hashing
---
* hash function
    * 임의의 길이의 데이터를 숫자로 맵핑하는 함수
    * 검색 속도를 높여줌
* hash value
    * 동일한 key 에 대해서는 같은 hash value 를 얻는다.
* hash() or `.__hash__(self)` 함수를 호출함
* hash() vs id()
    * id() 는 객체의 identity 를 반환함
    * 동일한 객체여도 다른 id 가 가능.
    * hash 는 동일한 객체이면 무조건 hash value 가 같다.
    * 같은 hash value 여도 값이 같지 않을 수 있다.
    
> 값의 uniqueness 를 체크하기 위해 hash()가 더 효율적이다.

# Hashable objects in Python
---
* 정의
    * lifetime 동안 hash value 가 바뀌지 않는 자료형
    * __hash__() and __eq__()
    * immutable 이면 hashable 타입
* 리스트랑 집합은 집합 아이템으로 넣을 수 없다.


```python
a = [10, 20, 30, 40, 30, 20, 10]
b = [2*i for i in range(51)]

a = set(a)
b = set(b)

c = a & b
d = a | b

print(a)
print(b)
print(c)
print(d)

c = list(c)
c.sort()
print(c)
```

```python
f = open('example.txt','r')
lines = f.readlines()
f.close()
lines = [line.rstrip() for line in lines]

word_dict = {}

for line in lines:
    for word in line.split():
        if word in word_dict:
            word_dict[word] += 1
        else:
            word_dict[word] = 1
            
for key in sorted(word_dict.keys()):
    if word_dict[key] >= 10:
        print(key, word_dict[key])
```
