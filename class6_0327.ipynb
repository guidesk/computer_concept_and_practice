# 실기 문제 1
---

```python
STR_INPUT = 'aabbaa'
LIST_INPUT = [1, 2, 3, 2, 1]
TUPPLE_INPUT = (2, 4, 3, 2)


def check_symmetry(input_sequence):
    # complete here
    for i in range(len(input_sequence)):
        if input_sequence[i] != input_sequence[len(input_sequence)-i-1]:
            return 0
    return 1       


def main():
    # complete here
    print(STR_INPUT, 'is '+ 'not '*(not check_symmetry(STR_INPUT)) + 'symmetric')
    print(LIST_INPUT, 'is '+ 'not '*(not check_symmetry(LIST_INPUT)) + 'symmetric')
    print(TUPPLE_INPUT, 'is '+ 'not '*(not check_symmetry(TUPPLE_INPUT)) + 'symmetric')

    
main()
```

# 실기 문제 2
---

```python
import random

NUMBER_OF_UNKNOWNS = 4

def generate_problem():
    # complete here
    # use random.randint()
    list_problem = []
    while len(list_problem)<NUMBER_OF_UNKNOWNS:
        list_problem.append(random.randint(0,9))
        if list_problem.count(list_problem[-1])>1:
            list_problem = list_problem[:-1]
    list_problem = ['%s'%l for l in list_problem]
    return list_problem


def guess_numbers():
    # complete here
    list_numbers = input("please enter %d numbers (0 <= n <= 9):"%NUMBER_OF_UNKNOWNS)
    list_numbers = [l for l in list_numbers]
    return list_numbers


def compare_two_numbers(target_numbers, numbers):
    # complete here
    exact_match = 0
    for i in range(NUMBER_OF_UNKNOWNS):
        if target_numbers[i] == numbers[i]:
            exact_match += 1
    inclusion = 0
    for i in range(NUMBER_OF_UNKNOWNS):
        if target_numbers.count(numbers[i])>0:
            inclusion += 1
    inclusion -= exact_match
    return exact_match, inclusion


def main():
    # complete here
    list_answer = generate_problem()
    for i in range(9):
        list_guess = guess_numbers()
        exact_match, inclusion = compare_two_numbers(list_answer,list_guess)
        print("---> exact_match: {}, inclusion: {}".format(exact_match,inclusion))
        if exact_match == NUMBER_OF_UNKNOWNS:
            break
    if i == 8:
        print("You failed to solve the problem.")
    else:
        print("Congratulations. The problem is solved.")

main()
```

# 실습 문제 3
---

```python
import random
import itertools


NUMBER_OF_POINTS = 5
MIN_COORD = 0
MAX_COORD = 20
INITIAL_DISTANCE = 99999999


def get_random_point():
    # complete here
    # use random.randint()
    coord_x = random.randint(MIN_COORD,MAX_COORD+1)
    coord_y = random.randint(MIN_COORD,MAX_COORD+1)
    return coord_x, coord_y
    

def generate_problem():
    # complete here
    coord_set = []
    for i in range(NUMBER_OF_POINTS):
        coord_set.append(get_random_point())
    return coord_set


def calc_distance(coord1, coord2):
    # complete here
    dist = ((coord1[0]-coord2[0])**2 + (coord1[1]-coord2[1])**2)**0.5
    return dist
    

def calc_sum_of_distances(points):
    # complete here
    dist_tot = 0
    for i in range(NUMBER_OF_POINTS):
        dist_tot += calc_distance(points[i-1],points[i])
    return dist_tot


def get_all_possible_orders():
    default_order = range(1, NUMBER_OF_POINTS)
    permutations = list(itertools.permutations(default_order))
    return permutations


def rearrange_points(points, tuple_of_indices):
    # complete here
    points_new = [points[0]]
    for i in range(len(tuple_of_indices)):
        n = tuple_of_indices[i]
        points_new.append(points[n])
    return points_new
    

def print_a_case(tuple_of_indices, distance):
    # complete here
    print("%s => distance = %.2f" %(tuple_of_indices,distance))


def main():
    # complete here
    points_problem = generate_problem()
    print(points_problem)
    possible_case = get_all_possible_orders()
    print(possible_case)
    best_case = [0]
    best_dist = [INITIAL_DISTANCE]
    
    for i in range(len(possible_case)):
        points_try = rearrange_points(points_problem, possible_case[i])
        dist_try = calc_sum_of_distances(points_try)
        print_a_case(possible_case[i],dist_try)
        
        best_case.append(possible_case[i])
        best_dist.append(dist_try)
        if best_dist[0] > best_dist[1]:
            best_case = [best_case[-1]]
            best_dist = [best_dist[-1]]
        else:
            best_case = [best_case[0]]
            best_dist = [best_dist[0]]
    print("This is the best case!")
    print_a_case(best_case[0],best_dist[0])
            

main()
```
